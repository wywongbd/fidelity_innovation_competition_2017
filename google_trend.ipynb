# Correlating Google Trend Data with Stock Volume

# Importing all dependencies

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from datetime import datetime
import gensim
```

# Fetch keywords similar or "close" to company name
Making use of a pre-trained Word2Vec model, a deep learning model to provide phrases that are semantically closest to input word, we obtained several keywords that are mostly related to HSBC, which are financial services, banking and investment.

```python
model = gensim.models.Word2Vec.load("wiki_word2vec_model") 
similar_words = model.most_similar("HSBC", topn=3)
searched_words = ["HSBC"] + [s[0] for s in similar_words]
print(searched_words)
```

# Retrieve Google Trends Data
We then retrieve trend data for all words in searched_words. The data retrieved represents the activity on a keyword on Google for the past 5 years. 

```python
word_trends = []
word_trends.append(pd.read_csv('HSBC.csv', header=1, names=['date', 'score'], parse_dates=[0]))
word_trends.append(pd.read_csv('FinancialServices.csv', header=1, names=['date', 'score'], parse_dates=[0]))
word_trends.append(pd.read_csv('Investment.csv', header=1, names=['date', 'score'], parse_dates=[0]))
word_trends.append(pd.read_csv('Banking.csv', header=1, names=['date', 'score'], parse_dates=[0]))
```

Below is the trend data for "HSBC" over the past 5 years. 

```python
word_trends[0]
```

Below is the visualized trends data for all 4 keywords.

```python
dates = [r[1]["date"].strftime("%Y-%m-%d") for r in word_trends[0].iterrows()]

plt.plot(dates, word_trends[0]["score"])
plt.plot(dates, word_trends[1]["score"])
plt.plot(dates, word_trends[2]["score"])
plt.plot(dates, word_trends[3]["score"])
plt.legend(["HSBC", "Financial Services", "Investment", "Banking"], loc=1)
plt.show()
```

# Calculate mean of trend data for all keywords
We now calculate the average trend by taking the mean value of trend data for all 4 keywords at all times. 

```python
averaged_trends = np.empty_like(word_trends[0]["score"])
for i, v in np.ndenumerate(averaged_trends):
    tmp = 0
    for w in word_trends:
        tmp += w.iloc[i]["score"]
    tmp /= len(word_trends)
    averaged_trends[i] = tmp

plt.plot(dates, averaged_trends)
plt.legend(["Averaged Trend"], loc=1)
plt.show()
```

# Loading volume data
Now we are done with the trend data. We now get the volume traded for HSBC for each of the days for which we have the trend data. This is to ensure that the two timeseries data are aligned.

```python
volume_hsbc = pd.read_csv("vol_hsbc_correct.csv", header=None).astype(float)
```

```python
volume_hsbc
```

# Overlay the averaged trend and volume traded for HSBC
Based on quick comparison, we realized that there is very little correlation between the averaged trend and the
volume   traded   by   HSBC

```python
plt.plot(dates, volume_hsbc[0])
plt.plot(dates, averaged_trends)
plt.legend(["Volume Traded", "Averaged Trends"], loc=1)
plt.show()
```

# Conclusion
One of the faults in our model is that keywords are static and stay the same over time. Changes in interest however take place due to events that impact the company. Since such events do not stay the same, keywords related to a company   need   to   be   changed.

For example, search hits for keywords like “HSBC Insider Trading” and “new financial regulations” could be a better representative for people’s interest in HSBC. However, this would hold only for a short period of time; more specifically, this would hold only as long as people care about the news. After this, events in the future will have a higher   impact   on   interest   in   HSBC.

Another factor that hindered our progress with this model was that it involved comparing time series data of the stock with time series data of the interest in the keyword. Due to the previously mentioned problem the model was unable to achieve an accurate level of prediction since there were only some places along the time series graph that the data presented correlation while at most other times the model could not detect any similarities. As a result we came to the conclusion that we would need to take an approach that would be more dynamic in terms of the keywords that were used   as   inputs   to   the   model,   to   reflect   the   change   in   interest   over   time.

Since changing events drive interest over time, we plan to train our Word2Vec model on different sources such as news and twitter to look for keywords representing events that impact people’s interest in the company. These keywords will change over time, making the model more dynamic. Instead of aggregating the mean of time series data, we will construct feature vectors for each timestamp to train our model to preserve the information within the data.
